```python
import matplotlib.pyplot as plt
import numpy as np


x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])


x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])
plt.suptitle('4 plots')
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x1,y1)
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x2, y2)
plt.title('Plot 4')
plt.show()
```
