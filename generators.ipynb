```python
import sys
def creator1():
  list =[]
  i=1
  while i<=200:
    list.append(i)
    i+=1
  return list
print(creator1())
z=sys.getsizeof(list)
print(z)
print([num+10 for num in creator1()])
```

```python
def creator2():
  i=1
  while i<=200:
    yield i
    i+=1
print(creator2())
x=creator2()
print(next(x))
print(next(x))
print(next(x))
print(list(x))
```
